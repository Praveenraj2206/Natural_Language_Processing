```python
import numpy as np
import pandas as pd
import textdistance
from collections import Counter
import re
```

```python
words=[]
with open('words.txt','r',encoding='utf-8')as f:
    data=f.read()
    data=data.lower()
    word=re.findall('\w+',data)
    words+=word
```

```python
word_freqdict=Counter(words)
# word_freqdict.most_common()
```

```python
# Probability(word)=Frequency/Total count of all words
# ex=['betty bought some butter,butter was bitter, to make the bitter butter better she put it into the batter to make the bitter butter,better butter']
# butter=6
# probalilty=6/26
```

```python
# total_words_freq=sum(word_freqdict.values())
# len(words)
```

```python
total_words_freq=sum(word_freqdict.values())

probs={}
for k in word_freqdict.keys():
    probs[k]=word_freqdict[k]/total_words_freq
```

```python
# probs
# print(probs)
```

```python
def predict(word):
    word=word.lower()
    if word in probs:
        print('the word is avilable',word)
    else:
        similarities=[1-(textdistance.Jaccard(qval=2)).distance(w,word) for w in word_freqdict.keys()]
        df=pd.DataFrame.from_dict(probs,orient='index').reset_index()
        df=df.rename(columns={'index':'word',0:'Prob'})
        df['Similarity']=similarities
        output=df.sort_values(['Similarity','Prob'],ascending=False).head(10)
        return(output)
predict('lov')
```
